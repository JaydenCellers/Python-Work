# Matplotlib Basics

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = x *2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.title('Title String')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 10)
plt.ylim(0,20)
plt.savefig('myfirstplot.png')
```
