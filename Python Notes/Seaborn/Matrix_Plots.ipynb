# Matrix Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df.head()
```

```python
df = df.set_index('Countries')
```

```python
df.head()
```

```python
sns.heatmap(df.drop('Life expectancy', axis=1), annot=True, cmap='viridis')
```

```python
sns.clustermap(df.drop('Life expectancy', axis=1), annot=True, cmap='viridis', col_cluster=False)
```
