# Categorical Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
sns.countplot(data=df, x='division', hue='level of education', palette='Set2')
```

```python
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, errorbar='sd')
```

```python
df2 = pd.read_csv('StudentsPerformance.csv')
```

```python
df2.head()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df2, y='writing score', x = 'parental level of education', hue='test preparation course')
```

```python
sns.violinplot(data=df2, y='writing score', x = 'parental level of education')
```

```python
sns.swarmplot(data=df2, x='math score', size=2, y ='gender')
```

```python
sns.boxenplot(data=df2, x='math score', y ='gender')
```
